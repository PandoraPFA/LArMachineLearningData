    Imports
</div>

```python
import numpy as np
import glob
import sys
```

<div class="alert alert-block alert-info" style="font-size: 18px;">
    Find files
</div>

```python
file_names = glob.glob(sys.path[0] + '/files/KalmanTraining_UVW_*.npz')
print(file_names)
```

<div class="alert alert-block alert-info" style="font-size: 18px;">
    Combine files
</div>

```python
x_tokens = []
y_tokens = []
is_contaminated = []
cluster_index = []

for file_name in file_names:
    print('file name:', file_name)
    
    with np.load(file_name) as data:   
        x_tokens.append(data["x_tokens"])
        y_tokens.append(data["y_tokens"])
        is_contaminated.append(data["is_contaminated"])
        cluster_index.append(data["cluster_index"])        
                  
x_tokens = np.concatenate(x_tokens, axis=0)
y_tokens = np.concatenate(y_tokens, axis=0)
is_contaminated = np.concatenate(is_contaminated, axis=0)
cluster_index = np.concatenate(cluster_index, axis=0)
```

```python
print(x_tokens.shape)
print(y_tokens.shape)
print(is_contaminated.shape)
print(cluster_index.shape)
```

<div class="alert alert-block alert-info" style="font-size: 18px;">
    Save files
</div>

```python
np.save(sys.path[0] + '/files/x_tokens_UVW.npy', x_tokens)
np.save(sys.path[0] + '/files/y_tokens_UVW.npy', y_tokens)
np.save(sys.path[0] + '/files/is_contaminated_UVW.npy', is_contaminated)
np.save(sys.path[0] + '/files/cluster_index_UVW.npy', cluster_index)
```
